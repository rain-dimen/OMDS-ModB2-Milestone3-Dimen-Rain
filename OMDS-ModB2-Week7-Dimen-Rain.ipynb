Below are the graphs from last week's weekly graph question.  Suppose you wanted to cause the viewer to examine the last three months of data first.  How could you use preattentive attributes to do that?  Suggest several possibilities.

ANSWER: You could add a box around the last three months in a contrasting color, you could make the last three data points big colored dots, you could thicken the line for the last three datapoints and make it a different color, or you could add arrows and annotations near the datapoints you want to highlight. 

```python
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
numdates = 12
np.random.seed(0)
time_series = 100 * (np.random.normal(size = numdates) / 5 + np.arange(numdates) / 16)
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Nitrate level (% of eutrophic value)")
```

```python
np.random.seed(1)
time_series = 100 * (1 - (np.random.normal(size = numdates) / 8 + np.arange(numdates) / 16))
datearray = pd.date_range(start='2023/01/01', end='2024/01/01', freq='M')
ax = sns.lineplot(x = datearray, y = time_series)
ax.set_xlabel("Year and month")
ax.set_ylabel("Phosphate level (% of eutrophic value)")
```

# 4. Storytelling With Data plot

Reproduce any graph of your choice in p. 99-119 of the Storytelling With Data book as best you can. (The first half of chapter four). You do not have to get the exact data values right, just the overall look and feel.

```python
import matplotlib.pyplot as plt

# Data
behaviors = [
    "Exercise 3+ times/week",
    "Eat 5+ servings of fruits/veggies/day",
    "Drink soda daily",
    "Smoke cigarettes",
    "Get annual physical",
    "Get <6 hours sleep/night"
]
percentages = [64, 28, 35, 18, 54, 23]

# Sort values for better visual structure (optional)
sorted_data = sorted(zip(percentages, behaviors))
sorted_percentages, sorted_behaviors = zip(*sorted_data)

# Plot
plt.figure(figsize=(10, 6))
bars = plt.barh(sorted_behaviors, sorted_percentages, color="#007acc")

# Emphasize one bar (e.g., "Smoke cigarettes")
highlight_index = sorted_behaviors.index("Smoke cigarettes")
bars[highlight_index].set_color("#d62728")

# Add labels to the end of each bar
for i, (bar, val) in enumerate(zip(bars, sorted_percentages)):
    plt.text(val + 1, bar.get_y() + bar.get_height()/2, f"{val}%", va='center')

# Style
plt.title("Health Behavior Participation Among U.S. Adults", fontsize=14)
plt.xlabel("% of Adults", fontsize=12)
plt.xlim(0, 80)
plt.gca().invert_yaxis()  # Highest at the top
plt.tight_layout()
plt.show()
```
