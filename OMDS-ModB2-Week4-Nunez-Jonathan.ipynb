```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import csv
from matplotlib.ticker import FuncFormatter

# === Load and clean general movies ===
df_movies = pd.read_csv("final_dataset_30kMovies.csv")
df_movies['Year'] = pd.to_numeric(df_movies['Year'], errors='coerce')
df_movies['budget'] = pd.to_numeric(df_movies['budget'], errors='coerce')
df_movies['grossWorldWide'] = pd.to_numeric(df_movies['grossWorldWide'], errors='coerce')

df_movies = df_movies[
    (df_movies['Year'] >= 2000) &
    (df_movies['Year'] <= 2023) &
    (df_movies['budget'].notna()) &
    (df_movies['budget'] > 0) &
    (df_movies['grossWorldWide'].notna()) &
    (df_movies['grossWorldWide'] > 0)
]

# Cap extreme budgets
df_movies = df_movies[df_movies['budget'] <= 5e8]

# Calculate ROI
df_movies['ROI'] = (df_movies['grossWorldWide'] / df_movies['budget']) - 1
df_movies = df_movies[df_movies['ROI'] <= 20]

# === Load and clean animated movies ===
df_animation = pd.read_csv(
    "Animation_Movies.csv",
    quotechar='"',
    quoting=csv.QUOTE_MINIMAL,
    encoding='utf-8'
)

df_animation['release_date'] = pd.to_datetime(df_animation['release_date'], errors='coerce')
df_animation['budget'] = pd.to_numeric(df_animation['budget'], errors='coerce')
df_animation['revenue'] = pd.to_numeric(df_animation['revenue'], errors='coerce')

df_animation = df_animation[
    (df_animation['release_date'].notna()) &
    (df_animation['budget'].notna()) &
    (df_animation['budget'] > 0) &
    (df_animation['revenue'].notna()) &
    (df_animation['revenue'] > 0)
]

df_animation['release_year'] = df_animation['release_date'].dt.year
df_animation = df_animation[
    (df_animation['release_year'] >= 2000) &
    (df_animation['release_year'] <= 2023)
]

# Cap extreme budgets
df_animation = df_animation[df_animation['budget'] <= 5e8]

# Calculate ROI
df_animation['ROI'] = (df_animation['revenue'] / df_animation['budget']) - 1
df_animation = df_animation[df_animation['ROI'] <= 20]

# === Plot: ROI vs Budget ===
plt.figure(figsize=(10, 6))

sns.regplot(
    x='budget', y='ROI',
    data=df_movies,
    scatter_kws={'alpha': 0.3},
    line_kws={'color': 'steelblue'},
    label='General Movies',
    color='steelblue'
)

sns.regplot(
    x='budget', y='ROI',
    data=df_animation,
    scatter_kws={'alpha': 0.4},
    line_kws={'color': 'orange'},
    label='Animated Movies',
    color='orange'
)

plt.title("ROI vs Budget (2000â€“2025, Capped at $500M)")
plt.xlabel("Budget")
plt.ylabel("Return on Investment")
plt.ylim(-1, 20)  # Show ROI up to 2000%

# Format x-axis in millions
formatter = FuncFormatter(lambda x, _: f'${x/1e6:.0f}M')
plt.gca().xaxis.set_major_formatter(formatter)

# Format y-axis as percentages
plt.gca().yaxis.set_major_formatter(FuncFormatter(lambda y, _: f'{y*100:.0f}%'))

plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Load NVDA stock data
df_nvda = pd.read_csv("NVDA.csv")

# Convert Date column to datetime
df_nvda['Date'] = pd.to_datetime(df_nvda['Date'], errors='coerce')

# Preview data
print(df_nvda.head())
print(df_nvda.dtypes)
```

```python
# --- NVDA: compute yearly average closing price
df_nvda['Year'] = df_nvda['Date'].dt.year
nvda_yearly = df_nvda.groupby('Year', as_index=False)['Close'].mean()

# --- Movies: count movies per year
df_movies_yearly = df_movies.groupby('Year', as_index=False).size().rename(columns={'size': 'Movie_Count'})
df_animation['release_year'] = pd.to_datetime(df_animation['release_date'], errors='coerce').dt.year
df_animation_yearly = df_animation.groupby('release_year', as_index=False).size().rename(columns={'release_year': 'Year', 'size': 'Animated_Count'})

# --- Merge all into one DataFrame
df_compare = pd.merge(nvda_yearly, df_movies_yearly, on='Year', how='left')
df_compare = pd.merge(df_compare, df_animation_yearly, on='Year', how='left')
df_compare[['Movie_Count', 'Animated_Count']] = df_compare[['Movie_Count', 'Animated_Count']].fillna(0)

# âœ… Limit to 2023 and earlier
df_compare = df_compare[df_compare['Year'] <= 2023]

# --- Plotting
import matplotlib.pyplot as plt

fig, ax1 = plt.subplots(figsize=(12, 6))

# NVDA stock price
ax1.plot(df_compare['Year'], df_compare['Close'], color='green', label='NVDA Avg Close ($)')
ax1.set_ylabel('NVDA Stock Price (Close)', color='green')
ax1.tick_params(axis='y', labelcolor='green')

# Movie counts
ax2 = ax1.twinx()
ax2.plot(df_compare['Year'], df_compare['Movie_Count'], label='General Movies', color='steelblue')
ax2.plot(df_compare['Year'], df_compare['Animated_Count'], label='Animated Movies', color='orange')
ax2.set_ylabel('Movie Count')
ax2.legend(loc='upper left')

plt.title("NVDA Stock vs Movie Output (2000â€“2023)")
plt.xlabel("Year")
plt.tight_layout()
plt.show()
```
