#EDA: US Accident Dataset

The US Accident dataset can be accessed here: https://www.kaggle.com/datasets/sobhanmoosavi/us-accidents

```python
US_Accidents_df = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Datasets/US_Accidents_March23_sampled_500k_Needed_Columns.csv')
US_Accidents_df
```

```python
US_Accidents_df.info()
```

```python
#What Start_Time and End_Time do not include the date
time_only = US_Accidents_df['Start_Time'].str.len() < 10
US_Accidents_df.drop(index=US_Accidents_df.loc[time_only].index, inplace=True)
```

```python
#Cast Start_Time String type to datetime64 type and reformat it to YYYY-MM-DD HH:MM:SS
try:
    US_Accidents_df["Start_Time"] = pd.to_datetime(US_Accidents_df["Start_Time"], format='mixed')
    print("Date was reformatted")
except ValueError as e:
    print(f"ValueError for AccountOpened: {e}")
```

```python
#Cast End_Time String type to datetime64 type and reformat it to  YYYY-MM-DD HH:MM:SS
try:
    US_Accidents_df["End_Time"] = pd.to_datetime(US_Accidents_df["End_Time"], format='mixed')
    print("Date was reformatted")
except ValueError as e:
    print(f"ValueError for AccountOpened: {e}")
```

```python
#Replace Null numeric values with mean
cols_to_fill = ['Temperature(F)', 'Wind_Chill(F)', 'Visibility(mi)', 'Wind_Speed(mph)', 'Precipitation(in)']

US_Accidents_df[cols_to_fill] = US_Accidents_df[cols_to_fill].fillna(
    US_Accidents_df[cols_to_fill].mean(numeric_only=True)
)
```

```python
#Find the day of the week in the Start_Time
US_Accidents_df['DayOfWeek'] = US_Accidents_df['Start_Time'].dt.day_name()
```

```python
#Replace Null categorical values with Unknown
US_Accidents_df[['Street', 'City', 'Zipcode']] = US_Accidents_df[['Street', 'City', 'Zipcode']].fillna('Unknown')
```

```python
#Replace Null categorical values with Unknown
US_Accidents_df[['Timezone', 'Wind_Direction', 'Weather_Condition']] = US_Accidents_df[['Timezone', 'Wind_Direction', 'Weather_Condition']].fillna('Unknown')
```

```python
#Replace Null categorical values with Unknown
US_Accidents_df[['Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight']] = US_Accidents_df[['Sunrise_Sunset', 'Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight']].fillna('Unknown')
```

```python
#Find the hour for each Start_Time
US_Accidents_df["Start_Hour"] = pd.to_datetime(US_Accidents_df["Start_Time"]).dt.hour
```

```python
US_Accidents_df.info()
```

```python
#Start Hour Histogram
sns.histplot(US_Accidents_df['Start_Hour'], bins = 24)
plt.title("Distribution of Accidents by Hour of Day")
plt.xlabel("Hour of Day (0 = Midnight, 23 = 11 PM)")
plt.ylabel("Number of Accidents")
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
subset = US_Accidents_df['Start_Hour'].dropna().sample(n=1000, random_state=42)
sns.swarmplot(y=subset, size=2)
plt.title("Distribution of Accident Start Times (Sample of 1,000)")
plt.ylabel("Hour of Day (023)")
plt.yticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
#Draw histograms and histogram variants for each feature or column.  (Swarm plot, kde plot, violin plot).
sns.kdeplot(US_Accidents_df['Start_Hour'])
plt.title("Density of Accidents by Hour of Day")
plt.xlabel("Hour of Day (0 = Midnight, 23 = 11 PM)")
plt.ylabel("Density")
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
sns.violinplot(US_Accidents_df['Start_Hour'])
plt.title("Distribution of Accidents by Hour of Day")
plt.ylabel("Hour of Day (0 = Midnight, 23 = 11 PM)")
plt.yticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
sns.histplot(US_Accidents_df['Severity'], bins=4, discrete=True)

plt.title("Distribution of Accident Severity")
plt.xlabel("Severity Level")
plt.ylabel("Number of Accidents")
plt.xticks([1, 2, 3, 4])
plt.tight_layout()
plt.show()
```

```python
US_Accidents_df['Start_Hour'] = US_Accidents_df['Start_Time'].dt.hour

hourly_severity_counts = US_Accidents_df.groupby(['Start_Hour', 'Severity']).size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
sns.barplot(data=hourly_severity_counts, x='Start_Hour', y='Count', hue='Severity', palette="viridis")

plt.title("Number of Accidents by Severity Level and Hour of Day")
plt.xlabel("Hour of Day (023)")
plt.ylabel("Accident Count")
plt.legend(title="Severity Level")
plt.tight_layout()
plt.show()
```

```python
weekday_order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']

US_Accidents_df['DayOfWeek'] = pd.Categorical(
    US_Accidents_df['DayOfWeek'],
    categories=weekday_order,
    ordered=True
)


plt.figure(figsize=(7, 4))
sns.histplot(data=US_Accidents_df, x='DayOfWeek', bins=7)
plt.title("Accidents by Day of Week")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
weekly_severity_counts = US_Accidents_df.groupby(['DayOfWeek', 'Severity']).size().reset_index(name='Count')

plt.figure(figsize=(12, 6))
sns.barplot(data=weekly_severity_counts, x='DayOfWeek', y='Count', hue='Severity', palette="viridis")

plt.title("Number of Accidents by Severity Level and Day of Week")
plt.xlabel("Day of Week")
plt.ylabel("Accident Count")
plt.legend(title="Severity Level")
plt.tight_layout()
plt.show()
```

```python
twilight_counts = US_Accidents_df[['Civil_Twilight', 'Nautical_Twilight', 'Astronomical_Twilight']].notna().sum()

sns.barplot(x=twilight_counts.index, y=twilight_counts.values)

plt.title("Number of Accidents by Twilight Type")
plt.xlabel("Twilight Type")
plt.ylabel("Count")
plt.tight_layout()
plt.show()
```

```python

top_20_counties = US_Accidents_df['County'].value_counts().head(20)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_20_counties.index, y=top_20_counties.values)

plt.title("Top 20 Counties by Accident Count")
plt.xlabel("County")
plt.ylabel("Accident Count")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
top_20_counties = US_Accidents_df['County'].value_counts().head(20).index

filtered = US_Accidents_df[US_Accidents_df['County'].isin(top_20_counties)]

county_to_label = (
    filtered[['County', 'State']]
    .drop_duplicates()
    .set_index('County')['State']
    .apply(lambda state: f"{state}")
    .to_dict()
)

county_to_label = {county: f"{county}, {state}" for county, state in county_to_label.items()}

county_counts = filtered['County'].value_counts().loc[top_20_counties]

plt.figure(figsize=(14, 6))
sns.barplot(x=county_counts.index, y=county_counts.values)

plt.xticks(
    ticks=range(len(county_counts)),
    labels=[county_to_label.get(c, c) for c in county_counts.index],
    rotation=45
)

plt.title("Top 20 Counties by Accident Count")
plt.xlabel("County (State label only)")
plt.ylabel("Accident Count")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_20_zipcodes = US_Accidents_df['Zipcode'].value_counts().head(20).index

filtered = US_Accidents_df[US_Accidents_df['Zipcode'].isin(top_20_zipcodes)]

label_map = (
    filtered[['Zipcode', 'City', 'State']]
    .dropna()
    .drop_duplicates('Zipcode')
    .set_index('Zipcode')
    .apply(lambda row: f"{row.name}, {row['City']}, {row['State']}", axis=1)
    .to_dict()
)

zipcode_counts = filtered['Zipcode'].value_counts().loc[top_20_zipcodes]

plt.figure(figsize=(14, 6))
sns.barplot(x=zipcode_counts.index, y=zipcode_counts.values)

plt.xticks(
    ticks=range(len(zipcode_counts)),
    labels=[label_map.get(zipcode, zipcode) for zipcode in zipcode_counts.index],
    rotation=45
)

plt.title("Top 20 Zipcodes by Accident Count")
plt.xlabel("Zipcode, City, State")
plt.ylabel("Accident Count")
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

top_20_cities = US_Accidents_df['City'].value_counts().head(20).index

filtered = US_Accidents_df[US_Accidents_df['City'].isin(top_20_cities)]


city_to_state = (
    filtered[['City', 'State']]
    .drop_duplicates('City')
    .set_index('City')['State']
    .to_dict()
)

city_counts = filtered['City'].value_counts().loc[top_20_cities]

city_labels = [f"{city} ({city_to_state.get(city, '')})" for city in city_counts.index]

plt.figure(figsize=(14, 6))
sns.barplot(x=city_counts.index, y=city_counts.values)

plt.xticks(ticks=range(len(city_labels)), labels=city_labels, rotation=45)

plt.title("Top 20 Cities by Accident Count")
plt.xlabel("City (State)")
plt.ylabel("Accident Count")
plt.tight_layout()
plt.show()
```
